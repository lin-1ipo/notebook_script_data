```python

# 素材

# 検索開始
for index, row in df.iterrows():
    filename = row['search_id'] + ".txt"
    # 出力上書き禁止
    if os.path.isfile(path + "/" + filename):
        print('既にファイルが存在します: ' + filename)
    else:
        opener = urllib.request.build_opener()
        opener.addheaders = [('User-agent', 'Mozilla/5.0')]

        # サイト内検索=True時のURL設定
        if not row['official_url'] and row['site_search']:
            #検索結果の読み込み（3件のみ表示）
            url_site = "https://www.google.co.jp/search?num=3&q=" + urllib.parse.quote(row['search_name'])          
            html = opener.open(url_site)
            bs_site = BeautifulSoup(html.read(), "lxml")
            # 0件ヒット
            if not bs_site.find_all("div",{"id":"res"})[0].get_text():
                continue
            # wikipediaが1位なら次の検索を使用する
            elif len(re.findall("wiki", bs_site.find_all("div",{"id":"res"})[0].cite.get_text())) == 0:
                # 日本語サイト判定
                if len(re.findall("/[^\./]*(ja)|(jp)[^\./]*$", bs_site.find_all("div",{"id":"res"})[0].cite.get_text())) > 0:
                    url_company = bs_site.find("div",{"id":"res"}).cite.get_text()
                else:
                    url_company = re.sub("(\.[a-z]+)/.+$", "\\1", bs_site.find("div",{"id":"res"}).cite.get_text())
            else:
                url_company = re.sub("/[^\.]+", "", bs_site.find("div",{"id":"res"}).find_all("cite")[1].get_text())
        else:
            url_company = row['official_url']

        # 検索実行（表示件数、サイト内検索かどうか）
        url_result = "https://www.google.co.jp/search?num=" + str(row['search_rows']) \
        + "&lr=lang_ja&q=" \
        + ("site:" + urllib.parse.quote(url_company) if row['site_search'] else urllib.parse.quote(row['search_name']))
        html = opener.open(url_result)
        bs_result = BeautifulSoup(html.read(), "lxml")
        
        # 検索結果：件数
        print(row['search_name'] + ' ' \
              + ("指定URLサイト内検索" if row['site_search'] else "ワード検索") + ' ' \
              + bs_result.find_all("div",{"id":"resultStats"})[0].get_text())
        
        # 検索結果ストック
        if len(bs_result.find_all("span",{"class":"st"})) > 0:
            result = "\n".join([re.sub("['\n\.]", "", x.get_text()) for x in bs_result.find_all("span",{"class":"st"})])
        else:
            continue

        #結果の書き込み
        fw = open(path + "/" + filename,'w', encoding='utf-8')
        fw.write(result)
        fw.close()
```
